```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/content/synthetic_fraud_dataset.csv"
df = pd.read_csv(file_path)

# Display basic information
print("Dataset Info:")
print(df.info())
print("\nBasic Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Display first few rows
print("\nFirst 5 Rows:")
print(df.head())

# Fill missing values with 0
df.fillna(0, inplace=True)

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Plot correlation heatmap
if not numeric_df.empty:
    plt.figure(figsize=(10, 6))
    sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title("Feature Correlation Heatmap")
    plt.show()
else:
    print("No numeric columns available for correlation heatmap.")

# Distribution of fraud cases (if 'fraud' column exists)
if 'fraud' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x='fraud', data=df)
    plt.title("Fraud Cases Distribution")
    plt.xlabel("Fraud (0: No, 1: Yes)")
    plt.ylabel("Count")
    plt.show()

# Pairplot for key numerical columns (only a subset for performance)
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()[:5]
sns.pairplot(df[numeric_cols])
plt.show()

# Save cleaned data (optional)
cleaned_file_path = "cleaned_fraud_dataset.csv"
if not df.empty:
    df.to_csv(cleaned_file_path, index=False)
    print(f"Cleaned dataset saved to {cleaned_file_path}")
else:
    print("Error: DataFrame is empty. Nothing to save.")
```

```python
# Load the dataset
file_path = "/content/synthetic_fraud_dataset.csv"
df = pd.read_csv(file_path)

# Standardize column names (remove spaces and convert to lowercase)
df.columns = df.columns.str.strip().str.lower()

# Check available columns
print("Columns in dataset:", df.columns)

# Check if 'fraud' column exists
if 'fraud' in df.columns:
    # Calculate fraud distribution
    fraud_counts = df['fraud'].value_counts()
    labels = ["Non-Fraudulent", "Fraudulent"]
    colors = ["skyblue", "salmon"]

    # Plot pie chart
    plt.figure(figsize=(6, 6))
    plt.pie(fraud_counts, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90, wedgeprops={'edgecolor': 'black'})
    plt.title("Fraudulent vs Non-Fraudulent Transactions")
    plt.show()
else:
    print("Error: 'fraud' column not found in the dataset. Available columns:", df.columns)

# Check if 'transaction_type' column exists
if 'transaction_type' in df.columns and 'fraud' in df.columns:
    # Calculate fraud rate per transaction type
    fraud_rates = df.groupby('transaction_type')['fraud'].mean().sort_values(ascending=False)

    # Plot fraud rates by transaction type
    plt.figure(figsize=(10, 6))
    sns.barplot(x=fraud_rates.index, y=fraud_rates.values, palette='coolwarm')
    plt.xlabel("Transaction Type")
    plt.ylabel("Fraud Rate")
    plt.title("Fraud Rate by Transaction Type")
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Error: 'transaction_type' column not found in the dataset.")
```

```python
# Identify potential fraud-related column
fraud_column_candidates = [col for col in df.columns if "fraud" in col.lower() or "label" in col.lower() or "is_fraud" in col.lower()]
if fraud_column_candidates:
    fraud_column = fraud_column_candidates[0]  # Take the first matching column
    print(f"Using column '{fraud_column}' for fraud analysis.")

    # 1. Pie Chart: Fraudulent vs. Non-Fraudulent Transactions
    fraud_counts = df[fraud_column].value_counts()
    labels = fraud_counts.index.astype(str)

    plt.figure(figsize=(6, 6))
    plt.pie(fraud_counts, labels=labels, autopct='%1.1f%%', colors=["skyblue", "salmon"], startangle=90, wedgeprops={'edgecolor': 'black'})
    plt.title("Fraudulent vs Non-Fraudulent Transactions")
    plt.show()
```

```python
    # 2. Bar Chart: Fraud Rate by Transaction Type
    if 'transaction_type' in df.columns:
        fraud_rates = df.groupby('transaction_type')[fraud_column].mean().sort_values(ascending=False)

        plt.figure(figsize=(10, 6))
        sns.barplot(x=fraud_rates.index, y=fraud_rates.values, palette='coolwarm')
        plt.xlabel("Transaction Type")
        plt.ylabel("Fraud Rate")
        plt.title("Fraud Rate by Transaction Type")
        plt.xticks(rotation=45)
        plt.show()
```

```python
# 3. Histogram: Transaction Amount Distribution
if 'transaction_amount' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df['transaction_amount'], bins=50, kde=True, color='blue')
    plt.xlabel("Transaction Amount")
    plt.ylabel("Frequency")
    plt.title("Transaction Amount Distribution")
    plt.show()
```

```python
    # 4. Box Plot: Fraud vs. Transaction Amount
    if 'transaction_amount' in df.columns:
        plt.figure(figsize=(8, 6))
        sns.boxplot(x=df[fraud_column].astype(str), y=df['transaction_amount'], palette='coolwarm')
        plt.xlabel("Fraud (0: No, 1: Yes)")
        plt.ylabel("Transaction Amount")
        plt.title("Transaction Amount by Fraud Status")
        plt.xticks(rotation=45)
        plt.show()
```

```python
    # 3. Joint Plot: Fraud Status vs. Transaction Amount
    if 'transaction_amount' in df.columns:
        sns.jointplot(x=df['transaction_amount'], y=df[fraud_column], kind='kde', fill=True, cmap='coolwarm')
        plt.suptitle("Joint Plot: Fraud Status vs. Transaction Amount", y=1.02)
        plt.show()
```

```python
    # 2. Hexbin Plot: Transaction Amount vs. Balance
    if 'transaction_amount' in df.columns and 'balance' in df.columns:
        plt.figure(figsize=(10, 6))
        plt.hexbin(df['transaction_amount'], df['balance'], gridsize=50, cmap='coolwarm', alpha=0.75)
        plt.xlabel("Transaction Amount")
        plt.ylabel("Balance")
        plt.title("Hexbin Plot: Transaction Amount vs. Balance")
        plt.colorbar(label='Count')
        plt.show()
```
